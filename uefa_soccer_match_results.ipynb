<b>Prognozy opracowane na podstawie danych dotyczących ilości strzelonych i straconych bramek przez analizowane drużyny w każdym meczu 
Ligi Mistrzów w sezonie 2021/2022 oraz w meczach Ligi Mistrzów sezonu 2022/2023, które odbyły się do dnia 16.04.2023. 
Dla każdego meczu za pomocą rozkładu Poissona obliczana jest wartość funkcji masy prawdopodobieństwa wyniku do 4 strzelonych bramek. 
Na podstawie otrzymanych wyników wybierana jest drużyna, która ma większą funkcję masy prawodopodobieństwa wygranej.</b>

```python
import numpy as np
import pandas as pd
from scipy.stats import poisson
```

```python
data = pd.read_excel(r"C:\Users\Klaudia\Desktop\data science\uefa\UEFAChampionsLeague2016-2022Data.xlsx", sheet_name="matches")
data_2023 = pd.read_csv(r"C:\Users\Klaudia\Desktop\data science\uefa\matches_champions_league_2022-23.csv")
```

```python
clubs = ["Real Madrid", "SL Benfica","Bayern München","AC Milan","Chelsea FC","SSC Napoli","Inter","Manchester City"]
clubs2023 = ["Real Madrid", "Benfica","Bayern Munich","Milan","Chelsea","Napoli","Inter","Manchester City"]
years = ["2021-2022","2020-2021"]
```

```python
#filtrowanie
data = data[data.HOME_TEAM.isin(clubs) | data.AWAY_TEAM.isin(clubs)]
data = data[data.SEASON.isin(years)]
```

```python
data
```

```python
data_2023 = data_2023[data_2023.team_name_home.isin(clubs2023) | data_2023.team_name_away.isin(clubs2023)]
```

```python
data_2023
```

<h3>Benfica vs. Inter</h3>

```python
Benefica_for=[]
for i in range(61):
    if data["HOME_TEAM"].iloc[i] == clubs[1]:
        Benefica_for.append(data["HOME_TEAM_SCORE"].iloc[i])
    if data_2023["team_name_home"].iloc[i] == clubs2023[1]:
        Benefica_for.append(data_2023["team_home_score"].iloc[i])
    if data["AWAY_TEAM"].iloc[i] == clubs[1]:
        Benefica_for.append(data["AWAY_TEAM_SCORE"].iloc[i])
    if data_2023["team_name_away"].iloc[i] == clubs2023[1]:
        Benefica_for.append(data_2023["team_away_score"].iloc[i])

Benefica_against=[]
for i in range(61):
    if data["HOME_TEAM"].iloc[i] == clubs[1]:
        Benefica_against.append(data["AWAY_TEAM_SCORE"].iloc[i])
    if data_2023["team_name_home"].iloc[i] == clubs2023[1]:
        Benefica_against.append(data_2023["team_away_score"].iloc[i])
    if data["AWAY_TEAM"].iloc[i] == clubs[1]:
        Benefica_against.append(data["HOME_TEAM_SCORE"].iloc[i])
    if data_2023["team_name_away"].iloc[i] == clubs2023[1]:
        Benefica_against.append(data_2023["team_home_score"].iloc[i])
        
Inter_for=[]
for i in range(61):
    if data["HOME_TEAM"].iloc[i] == clubs[6]:
        Inter_for.append(data["HOME_TEAM_SCORE"].iloc[i])
    if data_2023["team_name_home"].iloc[i] == clubs2023[6]:
        Inter_for.append(data_2023["team_home_score"].iloc[i])
    if data["AWAY_TEAM"].iloc[i] == clubs[6]:
        Inter_for.append(data["AWAY_TEAM_SCORE"].iloc[i])
    if data_2023["team_name_away"].iloc[i] == clubs2023[6]:
        Inter_for.append(data_2023["team_away_score"].iloc[i])
        
Inter_against=[]
for i in range(61):
    if data["HOME_TEAM"].iloc[i] == clubs[6]:
        Inter_against.append(data["AWAY_TEAM_SCORE"].iloc[i])
    if data_2023["team_name_home"].iloc[i] == clubs2023[6]:
        Inter_against.append(data_2023["team_away_score"].iloc[i])
    if data["AWAY_TEAM"].iloc[i] == clubs[6]:
        Inter_against.append(data["HOME_TEAM_SCORE"].iloc[i])
    if data_2023["team_name_away"].iloc[i] == clubs2023[6]:
        Inter_against.append(data_2023["team_home_score"].iloc[i])
        
Benefica_lambda = np.mean(Benefica_for)*np.mean(Benefica_against)
Inter_lambda = np.mean(Inter_for)*np.mean(Inter_against)
```

```python
Benefica_probs=[poisson.pmf(i, Benefica_lambda) for i in range(5)]
Inter_probs=[poisson.pmf(i,Inter_lambda)for i in range(5)]

match_probs=np.outer(Benefica_probs, Inter_probs)

for i in range(5):
    for j in range(5):
        print(f"Probability mass function of {i}-{j} match score: {match_probs[i,j]:.3f}")
```

<b>Z powyższej analizy wynika, że najbardziej prawodpodobny wynik meczu Benfica vs. Inter to 2-1. Wygrywa Benfica.</b>
